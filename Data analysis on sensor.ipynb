```python
## loading the dataset

import pandas as pd 
from functools import reduce
df=pd.read_csv('Nakuru_Sensor_Data_Feb_2023.csv.xls',delimiter=';')

df=df[['timestamp','value_type','value']]
print(df)
```

## Question E1:** Create a new collumn 'timestamp_new' from 'timestamp' rounded off to the second 

```python
df['timestamp'] = pd.to_datetime(df['timestamp']) # convert the timestamp column to datetime format
df['timestamp_new'] = df['timestamp'].dt.round('1s') # round off the timestamp to the second

print(df)
```

## Question E2:** Split the dataframe based on the 'value_type' collumn

```python
grouped = df.groupby('value_type')
df_temperature = df.groupby('value_type').get_group('temperature') # create a new dataframe with only temperature values
df_humidity = df.groupby('value_type').get_group('humidity') # create a new dataframe with only humidity values
df_P1 = df.groupby('value_type').get_group('P1') #You get the gist already
df_P2 = df.groupby('value_type').get_group('P2')
```

```python
df_temperature.head(10)
```

## Question E3:** Find the mean, mode and median of the humidity, temperature, P1 and P2 values

```python
# Calculate the mean, median, and standard deviation of temperature values
mu_temp = df_temperature['value'].mean()
med_temp = df_temperature['value'].median()
sd_temp = df_temperature['value'].std()
print("Temperature")
print("Mean = ",mu_temp, "Median = ", med_temp, "Standard Deviation = ", sd_temp)

# Calculate the mean, median, and standard deviation of humidity values
mu_humidity = df_humidity['value'].mean()
med_humidity = df_humidity['value'].median()
sd_humidity = df_humidity['value'].std()
print("Humidity")
print("Mean = ",mu_humidity, "Median = ", med_humidity, "Standard Deviation = ", sd_humidity)

mu_P1 = df_P1['value'].mean()
med_P1 = df_P1['value'].median()
sd_P1 = df_P1['value'].std()
print("P1")
print("Mean = ",mu_P1, "Median = ", med_P1, "Standard Deviation = ", sd_P1)

mu_P2 = df_P2['value'].mean()
med_P2 = df_P2['value'].median()
sd_P2 = df_P2['value'].std()
print("P2")
print("Mean = ",mu_P2, "Median = ", med_P2, "Standard Deviation = ", sd_P2)
```

## Question E4:** Merge the grouped dataframes into one dataframe and save it as newdata.csv

```python
# i got confused abiut this question seeing that I had created a variable called grouped earlier on.
# So that might be irrelavnat, thus i will merge the dataset based on Temp, Humidity, P1 & P2# Merge the datasets based on Temp, Humidity, P1 & P2
df_merged_temp_humd = pd.merge(df_temperature, df_humidity, on='timestamp_new', how = 'inner')
df_merged_P1 = pd.merge(df_merged_temp_humd, df_P1, on='timestamp_new', how = 'inner')
df_merged = pd.merge(df_merged_P1, df_P2, on='timestamp_new',how = 'inner', suffixes=('_left', '_right'))

#Renaming the columns
df_merged = df_merged.rename(columns={'value_type_x': 'value_type_temp', 'value_x' : 'value_temperature',
                        'value_type_y': 'value_type_humd',  'value_y' : 'value_humidity',
                        'value_type_left': 'value_type_P1',  'value_left' : 'value_P1',
                        'value_type_right': 'value_type_P2', 'value_right': 'value_P2'})

#print(df_merged)
df_merged.head()

from pathlib import Path  
filepath = Path('newdata.csv') 
filepath.parent.mkdir(parents=True, exist_ok=True)  
df_merged.to_csv("/Users/SBMCODED/Desktop/newdata.csv", index=False)  
```
